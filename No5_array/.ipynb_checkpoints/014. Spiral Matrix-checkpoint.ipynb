Given a matrix of m x n elements (m rows, n columns), return all elements of the matrix in spiral order.

    For example,
    Given the following matrix:

    [
     [ 1, 2, 3 ],
     [ 4, 5, 6 ],
     [ 7, 8, 9 ]
    ]
    You should return [1,2,3,6,9,8,7,4,5].

```python
class Solution:
    def __init__(self):
        self.result = []
        
    def spiralOrder(self, matrix):
        """
        :type matrix: List[List[int]]
        :rtype: List[int]
        """
        if len(matrix) != 0:
            self.n_row, self.n_col = len(matrix), len(matrix[0])
            i = 0
            while i <= (self.n_row-1) // 2 and i <= (self.n_col-1) // 2:
                if self.n_row - 2*i > 1 and self.n_col - 2*i > 1:
                    self.go_a_circle(matrix, i, i)
                elif self.n_row - 2*i == 1:
                    self.add_row(matrix, i, i)
                elif self.n_col - 2*i == 1:
                    self.add_col(matrix, i, i)
                
                i += 1
                    
        return self.result
        
    def go_a_circle(self, matrix, row_index, col_index):
        """
        :type matrix: List[List[int]]
        :type row_index: int
        :type col_index: int
        """
        n_row, n_col = len(matrix), len(matrix[0])
        
        row, col = row_index, col_index
        while(col < n_col - col_index - 1):
            self.result.append(matrix[row][col])
            col += 1
            
        while(row < n_row - row_index - 1):
            self.result.append(matrix[row][col])
            row += 1
            
        while(col > col_index):
            self.result.append(matrix[row][col])
            col -= 1
            
        while(row > row_index):
            self.result.append(matrix[row][col])
            row -= 1
            
    def add_row(self, matrix, row_index, col_index):
        self.result += matrix[row_index][col_index:len(matrix[0]) - col_index]
        
    def add_col(self, matrix, row_index, col_index):
        for row in range(row_index, len(matrix) - row_index):
            self.result.append(matrix[row][col_index])
            
```

```python
import unittest

class Test(unittest.TestCase):
    def __init__(self):
        pass
        
    def test_add_row(self):
        self.s = Solution()
        result = self.s.add_row([[ 1, 2, 3 ]], 0,0)
        expected = [ 1, 2, 3 ]
        self.assertListEqual(self.s.result, expected)

        
    def test_add_col(self):
        self.s = Solution()
        result = self.s.add_col([[1], [2], [3]], 0,0)
        expected = [ 1, 2, 3 ]
        self.assertListEqual(self.s.result, expected)
        
    def test_go_a_circle(self):
        self.s = Solution()
        result = self.s.go_a_circle([[1,2],[4, 3]], 0,0)
        expected = [ 1, 2, 3, 4 ]
        self.assertListEqual(self.s.result, expected)
        
    def test_spiralOrder(self):
        self.s = Solution()
        result = self.s.spiralOrder([[ 1, 2, 3 ], [ 4, 5, 6 ], [ 7, 8, 9 ]])
        expected = [1,2,3,6,9,8,7,4,5]
        self.assertListEqual(self.s.result, expected)
```

```python
t = Test()
t.test_add_row()
t.test_add_col()
t.test_go_a_circle()
t.test_spiralOrder()
```
