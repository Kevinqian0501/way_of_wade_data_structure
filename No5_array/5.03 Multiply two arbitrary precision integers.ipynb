```python
def multiply(input_1, input_2):
    """
        problem des:
            multiply two arbutrary precision integers
            
        hint:
            init a new array with length a+b
            each step: multiplication by one digit -> add -> shift to next digit
        
        Time: 
            O(n)
        Space:
            O(a + b)
            
    """
    
    result = [0]*(len(input_1) + len(input_2))
    
    sign = 1 if input_1[0]*input_2[0] > 0 else -1;
    
    input_1[0] = input_1[0] if input_1[0] > 0 else -input_1[0]
    input_2[0] = input_2[0] if input_2[0] > 0 else -input_2[0]
    
    
    #multiplication at ith digit
    carry = 0
    
    for i, v1 in enumerate(input_1[::-1]):
        for j, v2 in enumerate(input_2[::-1]):
            sum_ = v1 * v2 + result[i+j] + carry
            carry = sum_//10
            result[i+j] = sum_ % 10
        result[i+len(input_2)] += carry
        carry = 0
    
    for i in range(len(result)-1, -1, -1):
        v = result[i]
        if not v == 0:
            result[i] *= sign
            return result[:i+1][::-1]
        
    
    return [0]
    
```

```python
def test(input_, output_):
    
    result = multiply( input_[0], input_[1])
    
    for i, value in enumerate(result):
        if not value == output_[i]:
            print ("Wrong")
            print(result, output_)
            return -1
        
    print( "Correct")
    
    return 0
```
